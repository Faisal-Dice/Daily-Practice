```python
"""
Simple Linear Regression Model
Author: Faisal Hussain
Role: Data Scientist
Description: Predicts house prices based on area using Linear Regression.
"""

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample Dataset
data = {
    "Area": [500, 800, 1000, 1200, 1500, 1800, 2000],
    "Price": [100000, 150000, 200000, 250000, 300000, 350000, 400000]
}

df = pd.DataFrame(data)

# Features & Target
X = df[["Area"]]
y = df["Price"]

# Model Training
model = LinearRegression()
model.fit(X, y)

# Prediction
new_area = np.array([[1600]])
predicted_price = model.predict(new_area)

print(f"Predicted price for 1600 sq.ft house: ₹{int(predicted_price[0])}")

# Visualization
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Area (sq.ft)")
plt.ylabel("Price")
plt.title("House Price Prediction Model")
plt.show()
```
